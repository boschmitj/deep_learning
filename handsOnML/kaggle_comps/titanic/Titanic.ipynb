```python
from sklearn.base import BaseEstimator, TransformerMixin

class CategoricalImputer(BaseEstimator, TransformerMixin):
    def __init__(self):
        super().__init__()
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        X = X.copy()
        X["HomePlanet"] = X["HomePlanet"].fillna("Unknown")
        X["Destination"] = X["Destination"].fillna("Unknown")
        X["CabinDeck"] = X["Cabin"].str[0].fillna("Unknown")
        X = X.drop("Cabin", axis=1)
        return X
    
cat_imputer = CategoricalImputer()
titanic_cat_imputed = cat_imputer.transform(titanic_cat)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler()),
])

bool_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
])

cat_pipeline = Pipeline([
    ("imputer", CategoricalImputer()),
    ("encoder", OneHotEncoder(handle_unknown="ignore")),
])

```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(titanic_num)
cat_attribs = list(titanic_cat)
bool_attribs = list(titanic_bool)

drop_cols = ["Name", "PassengerId", "Transported"]
titanic_features = titanic.drop(columns=drop_cols)

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", cat_pipeline, cat_attribs),
    ("bool", bool_pipeline, bool_attribs),
])

titanic_prepared = full_pipeline.fit_transform(titanic)
```

```python
pd.DataFrame(titanic_prepared).head()
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(titanic_prepared, titanic_labels)
```

```python
from sklearn.metrics import accuracy_score

titanic_predictions = tree.predict(titanic_prepared)
tree_as = accuracy_score(titanic_labels, titanic_predictions)
tree_as
```

```python
from sklearn.model_selection import cross_val_score

cross_score = cross_val_score(tree, titanic_prepared, titanic_labels, cv=5, scoring="accuracy")
print("Corss-validation accuracy:", cross_score)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {"n_estimators": [30, 100], "max_features": [6, 10, 16, 20, 25]},
    {"bootstrap": [False], "n_estimators": [30, 100], "max_features": [6, 10, 16, 20, 25]},
]

grid_search = GridSearchCV(forest_cla, param_grid, cv=5, scoring="accuracy", return_train_score=True)

grid_search.fit(titanic_prepared, titanic_labels)
```
