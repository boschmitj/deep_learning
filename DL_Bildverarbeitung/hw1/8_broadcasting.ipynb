```python
import torch

a = torch.full((2, 3), 3)
b = torch.full((5, 1, 3), 3.)
print(a)
print(b)
```

```python
c = a + b
print(c)
print(c.shape)
```

```python
import numpy as np
def add(a, b):
    return a[:, np.newaxis] + b
```

```python
def add_with_unsqueeze(a, b):
    return a.unsqueeze(1) + b
```

```python
a = torch.arange(5)
b = torch.arange(3)
print(add(a, b))
print(add_with_unsqueeze(a, b))

print(a.shape)
print(a.unsqueeze(1).shape)
```

```python
def mult(a, b):
    return a.unsqueeze(3) * b.unsqueeze(1)
```

```python
g = torch.Generator().manual_seed(42)
a = torch.rand((3, 2, 5), generator=g)
b = torch.rand((3, 5, 4), generator=g)

print(mult(a,b))
```
