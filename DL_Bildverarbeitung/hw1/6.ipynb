```python
from PIL import Image
import numpy as np


image = Image.open("example_pic.jpg")
```

```python
# image_flattened = image.get_flattened_data()
# w = image.width
# h = image.height

# image_mat = np.array(image_flattened).reshape(176, 176)
# print(image_mat)

image_mat = np.array(image)
print(image_mat[80, :])
print(image_mat.dtype)
```

```python
import torch

image_tensor = torch.tensor(image_mat)
print(image_tensor)
```

```python
image_tensor_f32 = image_tensor.to(torch.float32)
```

```python
# def compute_avg_channel(channel: int, tensor):
#     return sum(image_tensor_f32[:, :, channel]) / (image_tensor_f32.shape[0] * image_tensor_f32.shape[1])

def compute_avg_channel(tensor: torch.Tensor):
    return tensor.mean((0, 1), keepdim=True)

def compute_std_channel(tensor: torch.Tensor):
    return tensor.std((0, 1), keepdim=True)
```

```python
avg_channels = compute_avg_channel(image_tensor_f32)
print(avg_channels)

std_channels = compute_std_channel(image_tensor_f32)
print(std_channels)
```
